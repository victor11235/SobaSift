```python
import pandas as pd
import numpy as np
import spacy
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS
from gensim.models import Phrases
from sklearn.feature_extraction.text import CountVectorizer
import tomotopy as tp

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import dataset_helper_fun.global_vars as global_vars
from dataset_helper_fun.utility import load_prof_email_dataset

```

```python

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load('en_core_web_sm')
# Create a list of stop words
stop_words = spacy.lang.en.stop_words.STOP_WORDS


def preprocess_text(text, bigram_model):
    # Tokenize text
    mytokens = nlp(text)

    # Lemmatization and convert to lower case
    mytokens = [word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens]

    # Remove stop words and non-alphabetic tokens
    mytokens = [word for word in mytokens if word not in stop_words and word.isalpha()]

    # Apply bigram model
    mytokens = bigram_model[mytokens]

    return mytokens


def load_dataset(file_path, file_encoding):
    # read emails
    df_emails, categories, prof_emails_labels, contents = load_prof_email_dataset(file_path, file_encoding)

    texts = df_emails['content'].apply(lambda x: x.split())
    bigram_model = Phrases(texts, min_count=5, threshold=0.5)

    df_emails['content'] = df_emails['content'].apply(lambda x: preprocess_text(x, bigram_model))

    frequency = pd.Series(' '.join(df_emails['content'].apply(lambda x: ' '.join(x))).split()).value_counts()
    rare_words = frequency[frequency == 1]
    common_words = frequency[frequency > 1000]  # adjust this for the actual corpus size

    df_emails['content'] = df_emails['content'].apply(
        lambda x: ' '.join([i for i in x if i not in common_words and i not in rare_words]))

    return df_emails, categories, prof_emails_labels


df, categories, prof_emails_labels = load_dataset('./datasets/chatgpt_prof_emails_500.csv', 'gbk')
df_val, categories_val, prof_emails_labels_val = load_dataset('./datasets/chatgpt_prof_emails_40.csv', 'gbk')

model = tp.SLDAModel(k=15, vars=['b'] * global_vars.NUM_EMAIL_LABEL_CATEGORIES)


# add data and train
content_list = df['content'].tolist()
content_list_val = df_val['content'].tolist()

for i in range(len(content_list)):
    words = content_list[i].split()
    labels = prof_emails_labels[i]
    model.add_doc(words=words, y=labels)
for i in range(0, 1000, 10):
    model.train(10)
    #print('Iteration: {}\tLog-likelihood: {}'.format(i, model.ll_per_word))

for k in range(model.k):
    print('Top 10 words of topic #{}'.format(k))
    print(model.get_topic_words(k, top_n=10))


def inference(content):
    doc = model.make_doc(words=content.split())
    model.infer(doc)
    return model.estimate(doc)


acc = []
label = 4
for j in range(len(df_val)):
    index = j
    if categories_val[index] == label:
        result = inference(content_list_val[index])
        print(result[label])
        acc.append(result[label])
        print(categories_val[index])
print(np.average(acc))
print(acc)


# ref: https://stackoverflow.com/questions/16945518/finding-the-index-of-the-value-which-is-the-min-or-max-in-python
def argmax(iterable):
    return max(enumerate(iterable), key=lambda x: x[1])[0]


```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams["figure.dpi"] = 220
plt.rcParams['text.usetex'] = True


# draw the confusion matrices
def draw_confusion_matrix_of_dataset(contents, content_labels, size):
    pred_labels = []
    for i in range(len(contents)):
        results = inference(contents[i])
        pred_label = argmax(results)
        pred_labels.append(pred_label)
    acc = np.average(np.array([pred_labels[i] == content_labels[i] for i in range(len(content_labels))], dtype=np.int32))
    print(f'accuracy: {acc}')
    cm = confusion_matrix(content_labels, pred_labels)
    plt.imshow(cm, cmap='Blues')
    plt.xlabel('Predicted', fontsize = size)
    plt.ylabel('True', fontsize = size)
    plt.title('Confusion Matrix', fontsize = 16)
    plt.colorbar()
    label_range = np.arange(global_vars.NUM_EMAIL_LABEL_CATEGORIES)
    label_strs = [str(label) for label in label_range]
    plt.xticks(label_range, label_strs)
    plt.yticks(label_range, label_strs)
    for i in range(len(cm)):
        for j in range(len(cm[0])):
            plt.text(j, i, str(cm[i, j]), ha='center', va='center', color='green')
    plt.show()
    

one_hot_labels = [np.argmax(row) for row in prof_emails_labels]
one_hot_labels_val = [np.argmax(row) for row in prof_emails_labels_val]
#draw_confusion_matrix_of_dataset(content_list, one_hot_labels)
draw_confusion_matrix_of_dataset(content_list_val, one_hot_labels_val, 14)
```
